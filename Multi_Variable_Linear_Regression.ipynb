# **Linear Regression 구현**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**1. 데이터 준비**

```python
x_train = np.array([ [2, 3, 1], [4, 8, 9], [6, 2, 7], [8, 4, 6] ])
```

```python
y_train = np.array([[4], [7], [5], [8]])
```

```python
n_samples = x_train.shape[0]
```

```python
n_features = x_train.shape[1]
```

**2. 가설 함수 설정**

 - f(x) = w1x1 + w2x2 + w3x3 + b



```python
W = np.ones((3,1))      # 1로 초기화 weight parameters
```

```python
b = np.ones((1,1))
```

```python
def f(X):
    # return x[0]*W[0] + x[1]*W[1] + x[2]*W[2] + b 
    return np.matmul(X, W) + b
```

```python
y_pred = f(x_train)
```

**3. Cost함수 정의**

```python
def loss_function(y_pred, y_train):
    n_samples = y_train.shape[0]
    deviation = y_pred - y_train
    deviation_squared = deviation**2
    # print(deviation_squared)
    loss = (1/n_samples) * deviation_squared.sum()

    return loss
```

```python
loss_function(y_pred, y_train)
```

```python
def gradient(x_train, y_train, y_pred):
    n_samples = y_train.shape[0]
    dw = (1/n_samples) * np.matmul(x_train.transpose(), (y_pred - y_train))
    db = (1/n_samples) * np.sum(y_pred - y_train, axis=0, keepdims=True)
    print(dw, db)

    return dw, db
```

```python
dw = gradient(x_train, y_train, y_pred)
db = gradient(x_train, y_train, y_pred)
```

```python
def gradient_descent(W, b, learning_rate, dw, db):
    W = W - learning_rate * dw
    b = b - learning_rate * db
    print(W)

    return W, b
```

```python
gradient_descent(W, b, 0.01, dw, db)
```

```python
losses = []

for num_iter in range(100):
    print("x_train:", x_train, x_train.shape)
    y_pred = f(x_train)
    print("y_pred:", y_pred, y_pred.shape)
    loss = loss_function(y_pred, y_train)
    losses.append(loss)

    dw, db = gradient(x_train, y_train, y_pred)
    W, b = gradient_descent(W, b, 0.01, dw, db)

    print(W)
    print(loss)

costs = np.squeeze(losses)    
plt.plot(losses)
plt.ylabel('Mean Squred Error')
plt.xlabel('The number of iteration')
plt.title("Multi-variable Linear Regression")
plt.show()
```
